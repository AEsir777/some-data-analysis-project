```python
import pandas as pd
import numpy as np
import os
import requests
from IPython.display import Image as Ima
from IPython.display import clear_output, display
from PIL import Image as Im
import joblib
import matplotlib.pyplot as plt
import json
import base64
import shutil
import warnings
```

```python
clusters1 = np.load('ai-village-ctf/math/clusters1.npy')
clusters2 = np.load('ai-village-ctf/math/clusters2.npy')
clusters3 = np.load('ai-village-ctf/math/clusters3.npy')
```

```python
clusters1.shape
```

```python
import sklearn
import sklearn.metrics as sm

from sklearn.cluster import AgglomerativeClustering

import scipy
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
z1 = linkage(clusters1, 'ward')
```

```python
dendrogram(z1, truncate_mode = 'lastp', p = 12, leaf_rotation = 45., 
          leaf_font_size = 15, show_contracted = True)
plt.title('Dendrogram')
plt.xlabel('Cluster')
plt.ylabel('Distance')

plt.axhline(y = 25, c = 'gray') # 5 clusters
plt.axhline(y = 150, c= 'gray') # 3 clusters

plt.show()
```

```python
z2 = linkage(clusters2, 'ward')
dendrogram(z2, truncate_mode = 'lastp', p = 12, leaf_rotation = 45., 
          leaf_font_size = 15, show_contracted = True)
plt.title('Dendrogram')
plt.xlabel('Cluster')
plt.ylabel('Distance')

plt.axhline(y = 25, c = 'gray') # 2 cluasters
plt.axhline(y = 150, c= 'gray') # 2 clusters

plt.show()
```

```python
z3 = linkage(clusters3, 'ward')
dendrogram(z3, truncate_mode = 'lastp', p = 12, leaf_rotation = 45., 
          leaf_font_size = 15, show_contracted = True)
plt.title('Dendrogram')
plt.xlabel('Cluster')
plt.ylabel('Distance')

plt.axhline(y = 25, c = 'gray') # 3 clusters
plt.axhline(y = 150, c= 'gray')

plt.show()
```

## Math 2

```python
dim1 = np.load('ai-village-ctf/math/first_dim1.npy')
dim2 = np.load('ai-village-ctf/math/first_dim2.npy')
dim3 = np.load('ai-village-ctf/math/first_dim3.npy')
dim1
```

```python
from sklearn.decomposition import FactorAnalysis
from sklearn import decomposition
from sklearn.decomposition import PCA
```

```python
factor = FactorAnalysis().fit(dim1)

df1 = pd.DataFrame(factor.components_)
df1
```

```python
pca1 = decomposition.PCA()
pca1.fit_transform(dim1)

pca1.explained_variance_ratio_
# 3 dim
```

```python
factor2 = FactorAnalysis().fit(dim2)

df2 = pd.DataFrame(factor2.components_)
df2
```

```python
pca2 = decomposition.PCA()
pca2.fit_transform(dim2)

pca2.explained_variance_ratio_
# 5 dim
```

```python
factor3 = FactorAnalysis().fit(dim3)

df3 = pd.DataFrame(factor3.components_)
df3
```

```python
pca3 = decomposition.PCA()
pca3.fit_transform(dim3)

pca3.explained_variance_ratio_
# 4 dim
```

## Math 4

```python
from sklearn.cluster import KMeans
```

```python
clt = KMeans(n_clusters = 5, random_state = 20902490) # 5 clusters
clt.fit(clusters1)
```

```python
clusters1.shape
```

```python
clt.labels_
```

```python
print((clt.labels_ == 0).sum())
print((clt.labels_ == 1).sum())
print((clt.labels_ == 2).sum())
print((clt.labels_ == 3).sum())
print((clt.labels_ == 4).sum())
```

```python
c1 = clusters1[clt.labels_ == 0]
c2 = clusters1[clt.labels_ == 1]
c3 = clusters1[clt.labels_ == 2]
c4 = clusters1[clt.labels_ == 3]
c5 = clusters1[clt.labels_ == 4]
c5.shape
```

```python
pca = decomposition.PCA()
pca.fit_transform(c1)

pca.explained_variance_ratio_ # 4
```

```python
pca.fit_transform(c2)
pca.explained_variance_ratio_ # 2
```

```python
pca.fit_transform(c3)
pca.explained_variance_ratio_ # 1
```

```python
pca.fit_transform(c4)
pca.explained_variance_ratio_ # 5
```

```python
pca.fit_transform(c5)
pca.explained_variance_ratio_ # 3
```
